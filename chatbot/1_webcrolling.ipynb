# 웹 크롤링

네이버 영화의 리뷰를 크롤링

```python
from selenium import webdriver
import chromedriver_autoinstaller
from tqdm import tqdm
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
chromedriver_autoinstaller.install(cwd=True)
driver=webdriver.Chrome()
```

```python
dic={}
```

영화 이름과 리뷰를 담을 dic이라는 이름의 빈 딕셔너리를 생성한다.

```python
code=178000
for i in tqdm(range(175000,code)):
    url_movie='https://movie.naver.com/movie/bi/mi/point.naver?code='+str(i)
    driver.get(url_movie)
    try:
        html=driver.page_source
        bs=BeautifulSoup(html,'html.parser')
        bss=bs.findAll('p', {'class':'tx_report'})
        if bss==[]:
            pass
        else:
            sc=bs.find('span', {'class':'st_on'}).get_text()
            if int(re.findall('[0-9].', sc)[0][0]) > 6:
                driver.get('https://movie.naver.com/movie/bi/mi/basic.naver?code='+str(i))
                html=driver.page_source
                bs1=BeautifulSoup(html,'html.parser')
                intro=bs1.findAll('p', {'class':'con_tx'})[0].text
                name=bs.title.text.rstrip(': 네이버 영화')
                tex1=bss[0].text
                dic[name]=tex1,intro
    except:
        pass
```

BeautifulSoup을 이용하여 크롤링한 후, 리뷰 부분을 bss에 저장한다.
리뷰가 없는 영화는 생략하고, 리뷰가 있는 영화에 한해 크롤링을 진행한다.

영화를 추천하는 프로그램이므로, 평점을 고려해야 한다고 생각하여 평점이 10점만점에 7점이 넘는 영화만 반영하도록 하였다. 

영화의 이름을 dic의 key에, 리뷰와 소개글을 dic의 value에 저장하였다.

시간상의 문제로 최근에 나온 영화를 중심으로 크롤링 하였고 2017년부터 2022년까지를 중심으로 하였다.(코드 기준 178000~230000/ 약 5만개의 영화)

```python
movie=pd.DataFrame(columns=['movie_name','review_text'])
movie.movie_name=dic.keys()
movie.review_text=dic.values()

movie
```

데이터 프레임으로 만들어 movie에 할당하고 이를 csv파일로 저장한다.

```python
movie.to_csv('mv.csv', encoding='utf-8-sig',index=False, header=False)
```
